```python
def prepare_dataframe(res_dict, msg, m_lst = ['F1-Weighted'], sota=''):
    for d, exp_dict in res_dict.items():
        out_dict = {}
        # print(d, len(exp_dict['Org']), 'Org')
        if 'Org' in exp_dict:
            for splt_item in exp_dict['Org']:
                for m in m_lst:
                    v = splt_item[2]['Org'][0][m]
                    # print(v, splt_item)
                    k = msg + '.' + m[:4] + '.' + 'Org'
                    if not k in out_dict:
                        out_dict[k] = []    
                    out_dict[k].append(v)
        for CF, splt_items in exp_dict['CF'].items():
            for splt_item in splt_items:
                # print(CF, splt_item)
                for m in m_lst:
                    v = splt_item[2]['CF'][0][m]
                    k = msg + '.' + m[:4] + '.' + sota + 'CF-' + CF[:3]
                    if not k in out_dict:
                        out_dict[k] = []
                    out_dict[k].append(v)
        _res_df = pd.DataFrame(out_dict)
        _res_df.loc['Mean'] = _res_df.mean()
        _dict = _res_df.loc['Mean'].to_dict()
        out_dict = {}
        out_dict['Domain'] = d
        if 'Org' in exp_dict:
            k = msg + '.' + m[:4] + '.' + 'Org'
            nk = '$' + msg + '.' + m[:2] + '_' + m[3] + '$'
            out_dict[nk] = f'{round(_dict[k], 3):.3f}'
            del _dict[k]
        for i, (_k, v) in enumerate(sorted(_dict.items(), key=lambda x:x[1], reverse=True)[:1]):
            k = msg + '.' + m[:4] + '.' + 'CF'
            nk = '$' + msg + '.' + m[:2] + '_' + m[3] + '.' + sota + 'cf' + '$'
            # if i == 1:
            #     nk += ' 2nd'
            # nk += ' Best'
            out_dict[nk] = f'{round(v, 3):.3f}' + '$_{cf=' + _k[-3:]+'}$'
        yield d, _res_df, out_dict
```
